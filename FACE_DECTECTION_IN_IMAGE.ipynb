```python
#Importing the necessary modules and reading the image
import cv2
imagepath = r"C:\Users\gayat\OneDrive\Desktop\input_image.jpg"
img= cv2.imread(imagepath)
```

```python
#Print the dimensions of the array
img.shape
#we can see that the output has 3 dimension output , thats because (height , weighth and channels ) as this is an colour image it has BGR colours so we have to change that to gray scale
```

```python
#convert the image to gray scales before performing the face detection
gray_image = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
#as this is a gray scale image it no longer has the third channel
gray_image.shape
```

```python
#haarcascade_frontalface_default.xml is a built in classifier is designed specifically for detecting frontal faces in visual inputs
face_classifier = cv2.CascadeClassifier( cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)
#detectMultiscale() is used to identify faces of different sizes in the input image
#gray_image is the gray scale image we created previously
#scalefactor it is use dto scale down the size of the input image and it makes it easier for the algo to detect the larger faces,
#Here the scale factor is 1-1 indicating that we want to reduce the image by 10%
#min neighbors indicates how many times a particular feature point should be detected within an image.
#The minimum size parameter specifies the smallest possible object you'd like to detect (in pixels) The model will ignore faces that are smaller than the minimum size specified.
```

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)

img_rgb = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt



plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
```
